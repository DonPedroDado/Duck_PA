```python
from pytesseract import pytesseract
from PIL import Image
import os
import re

pytesseract.tesseract_cmd = 'tesseract'

input_png = os.path.expanduser("test5.png")
output_txt = os.path.expanduser("/Users/donpedrodado/Documents/opt/Duck_TAI/output10.txt")

def clean_ocr_text(text):
    text = re.sub(r'/(\d+)', r'√\1', text)
    text = re.sub(r'(\w+)\s+(\w+)', r'\1/\2', text)
    text = re.sub(r'(Examples of rational numbers:)', r'\1\n- ', text)
    text = re.sub(r'(Examples of irrational numbers:)', r'\1\n- ', text)
    text = re.sub(r';', r'\n-', text)
    text = re.sub(r'0,\d', lambda match: match.group(0).replace(',', '.'), text)
    text = re.sub(r'(DEFINITION:.*?)\s', r'\1\n', text)
    text = re.sub(r'(NOTE.*?)\s', r'\1\n', text)
    text = re.sub(r'\s+', ' ', text).strip()
    text = re.sub(r'\s\.\s', '. ', text)
    return text


print("Opening PNG image...")
image = Image.open(input_png)

print("Performing OCR...")
ocr_text = pytesseract.image_to_string(image, lang='eng')

print("Cleaning OCR text...")
cleaned_text = clean_ocr_text(ocr_text)

print("Saving cleaned text to file...")
with open(output_txt, "w", encoding="utf-8") as f:
    f.write(cleaned_text)

print(f"OCR and cleaning complete! Text saved to: {output_txt}")
```

```python
import pytesseract
from PIL import Image
import re
import os

def extract_text_from_image(image_path):
    try:
        img = Image.open(image_path)
        text = pytesseract.image_to_string(img, lang='eng')
        return text
    except Exception as e:
        return f"Error during OCR processing: {str(e)}"

def clean_text(text):
    text = re.sub(r'/(\d+)', r'√\1', text)
    text = re.sub(r'\s+', ' ', text).strip()
    text = re.sub(r';', r'\n- ', text)
    text = re.sub(r'(\d),(\d)', r'\1.\2', text)
    text = re.sub(r'([A-Z ]+):\s*', r'\1:\n', text)
    text = re.sub(r'\s+\.\s+', '. ', text)
    text = re.sub(r'\.{2,}', '.', text)
    return text

def process_image(image_path):
    extracted_text = extract_text_from_image(image_path)
    print("\n--- Extracted Text ---\n")
    print(extracted_text)
    cleaned_text = clean_text(extracted_text)
    print("\n--- Cleaned Text ---\n")
    print(cleaned_text)

if __name__ == "__main__":
    image_path = os.path.expanduser("test5.png")
    process_image(image_path)
```
